# Nettoyage de la base de données

### Certains individus ont plusieurs fois un même numéro de déplacement. Nous allons supprimer ces individus car ils sont négligeables devant la totalité des individus.

```python
import pandas as pd
import numpy as np
import math
!pip install openpyxl
```

```python
data = pd.read_excel("data/EMP_deplacements.xlsx")
```

```python
data.head()
```

```python
# Trouver les valeurs de "IDENT_IND" avec plusieurs occurrences de "num_dep_V"
duplicate_num_ind = data[data.duplicated(subset=['IDENT_IND', 'num_dep_V'], keep=False)]['IDENT_IND']

# Supprimer les lignes correspondant à ces valeurs de "num_ind"
data = data[~data['IDENT_IND'].isin(duplicate_num_ind)]
```

```python
data.to_excel("data/data.xlsx")
```
