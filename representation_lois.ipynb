Ce fichier a pour but de représenter graphiquement les différentes densités des lois à simuler.

### 1. Loi du temps de trajet.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Remplacez ces distributions et paramètres par les vôtres
dist1, params1 = stats.lognorm, (0.831523, -0.046935, 13.901157)  # matin
dist2, params2 = stats.lognorm, (0.740185, -0.255895, 12.666298)   # 11_14
dist3, params3 = stats.lognorm, (0.765396, -0.398847, 15.391876) # 14_17
dist4, params4 = stats.invgauss, (0.557836, -2.033562, 37.452790) # soir

# Créer un ensemble de points x pour lesquels calculer la PDF
x = np.linspace(0, 200, 1000)

# Calculer les PDF pour chaque distribution
pdf1 = dist1.pdf(x, *params1)
pdf2 = dist2.pdf(x, *params2)
pdf3 = dist3.pdf(x, *params3)
pdf4 = dist4.pdf(x, *params4)

# Tracer les PDFs
plt.figure(figsize=(10, 6))
plt.plot(x, pdf1, label='Distribution 0h-11h')
plt.plot(x, pdf2, label='Distribution 11h-14h')
plt.plot(x, pdf3, label='Distribution 14h-17h')
plt.plot(x, pdf4, label='Distribution 17h-00H')

plt.title('Comparaison de densités pour la loi du temps de trajet')
plt.xlabel('Valeur')
plt.ylabel('Densité de probabilité')
plt.legend()
plt.grid(True)
plt.show()
```

### 2. Loi du premier départ.
